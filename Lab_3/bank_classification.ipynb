# Load the data

```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/albanda/CE888/master/lab3/bank-additional-full.csv', delimiter = ";")
df
```

# Data preprocessing

```python
from sklearn.preprocessing import LabelEncoder

df[['y']] = df[['y']].apply(LabelEncoder().fit_transform)
df.head()
```

```python
df = pd.get_dummies(df)
df.head()
```

```python
df = df.drop(columns=['duration'])
df.head()
```

```python
df.isnull().any().sum()
```

```python
import seaborn as sns

sns.displot(df, x="y", bins=2)
```

# Training the model

```python
from sklearn.model_selection import train_test_split

X_df = df.drop(columns=['y'])
Y_df = df['y']

x, x_test, y, y_test = train_test_split(
    X_df.values,
    Y_df.values,
    test_size=0.2,
    random_state=50,
    stratify=Y_df.values
)
```

```python
from sklearn.metrics import f1_score, make_scorer

scorer = make_scorer(f1_score)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

clf = RandomForestClassifier(max_depth=3, n_estimators=100)
scores = cross_val_score(clf, x, y, cv=10, scoring=scorer)
print(f'ACC: {scores.mean():.2%} +/- {scores.std():.2%}')
```

```python
from sklearn.dummy import DummyClassifier

dummy_clf = DummyClassifier(strategy='stratified')
scores = cross_val_score(dummy_clf, x, y, cv=10, scoring=scorer)
print(f'ACC: {scores.mean():.2%} +/- {scores.std():.2%}')
```

```python
parameters = {
    'n_estimators': (10, 100, 1000),
    'criterion': ('gini', 'entropy'),
    'max_depth': (2, 3, 6),
}
```

```python
from sklearn.model_selection import GridSearchCV

clf = GridSearchCV(RandomForestClassifier(), parameters, cv=10, scoring=scorer)
clf.fit(x, y)
```

```python
print(clf.best_params_)
f'ACC: {clf.best_score_:.2%}'
```

```python
import numpy as np

best_estimator = clf.best_estimator_
features = X_df.columns

importances = best_estimator.feature_importances_[:3]
std = np.std([tree.feature_importances_ for tree in best_estimator.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]
print(indices)

# Print the feature ranking
print("Feature ranking:")

for f in range(3):
    print("%d. %s (%f)" % (f + 1, features[indices[f]],  importances[indices[f]]))
```

```python
import matplotlib.pyplot as plt

# Plot the feature importances of the forest
fig = plt.figure()
plt.title("Feature importances")
plt.bar(range(3), importances[indices],
       color="r", yerr=std[indices], align="center")
plt.xticks(range(3), np.array(features)[indices])
plt.xlim([-1, 3])
plt.ylim([0, None])

fig.set_size_inches(15, 8)
```

```python
best_estimator.score(x_test, y_test)
```
